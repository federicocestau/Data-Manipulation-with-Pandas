```python
 response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
Price = soup.find_all("b",class_="sale-price nc_PIPprice")
Price
```

```python
Features = soup.find_all("div",class_="prod-detail-content nc_tab_content")
Features
```
