```python
def my_fun(x, var1, var2, var3):
    #print (x)
    mask = (x[var1]-x[var2])>0
    x.loc[mask, var3] = x[var1]-x[var2]
    x.loc[~mask, var3] = 0
    return x
```

```python
def sort_words(text):
    words = text.split(' ')
    unique_words = set(words)
    word_counts = []
    for word in unique_words:
        word_counts.append((words.count(word), word))
    word_counts.sort(reverse=True)
    sorted = []
    for count, word in word_counts:
        sorted.append(word)
    return sorted
```
