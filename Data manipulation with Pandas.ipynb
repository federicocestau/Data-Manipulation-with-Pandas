Joins using Merge. 

1)one-to-one-joins

```python
df1 = pd.DataFrame({'employee': ['Bob','Jake','Lisa','Sue'],
               'group': ['Accounting','Engineering','Engineering','HR']})
df1
```

```python
df2 = pd.DataFrame({'Person': ['Jake','Bob','Lisa','Sue'],
               'hire_date': [2004,2010,2015,2018]})
df2
```

```python
df3 = df1.merge(df2, left_on='employee', right_on='Person',how='inner')
df3
```

```python
df1.merge(df2, left_on='employee', right_on='Person',how='outer')
```

2) One to many joins. One of the two key columns contain duplicate values. 

```python
df4 = pd.DataFrame({'group' : ['Accounting','Engineering','HR'],
                   'supervisor' : ['Carly','Guido','Steve']})
```

```python
df5 = pd.merge(df3,df4)
df5
```

3) both of the two key columns contains duplicate values. 

```python
df6 = pd.DataFrame({'group' : ['Accounting','Accounting','Engineering','Engineering','HR','HR'],
                   'skills' : ['math','spreadsheets','coding','linux','spreadsheets', 'organization']})
df6
```

```python
df7 = df1.merge(df6,left_on='group', right_on='group',how='outer')
df7
```

```python
df8 = df3.merge(df6,left_on='group', right_on='group',how='outer').drop('Person',axis=1)
df8
```

The Pandas library provides a function called resample() on the Series and DataFrame objects. This can be used to group records when downsampling and making space for new observations when upsampling.

How to Use “OR” Operator in Pandas (With Examples)

For example, you can use the following basic syntax to filter for rows in a pandas DataFrame that satisfy condition 1 or condition 2:

df[(condition1) | (condition2)]

```python
df = pd.DataFrame({'team': ['A', 'A', 'B', 'B', 'B', 'B', 'C', 'C'],
                   'points': [25, 12, 15, 14, 19, 23, 25, 29],
                   'assists': [5, 7, 7, 9, 12, 9, 9, 4],
                   'rebounds': [11, 8, 10, 6, 6, 5, 9, 12]})

#view DataFrame
df
```

```python
#filter rows where points > 20 or assists = 9
df[(df.points > 20) | (df.assists == 9)]
```

```python
#create DataFrame
df = pd.DataFrame({'team': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
                   'position': ['G', 'G', 'F', 'F', 'C', 'F', 'C', 'C'],
                   'conference': ['W', 'W', 'W', 'W', 'E', 'E', 'E', 'E'],
                   'points': [11, 8, 10, 6, 6, 5, 9, 12]})

#view DataFrame
df
```

```python
#filter rows based on string values
df[(df.team == 'H') | (df.position == 'G') | (df.position == 'F') | (df.conference == 'F')]
```
