# loc function

loc is a way to filter over the index, row section,and you can choose the columns tyou want to see but not in a range way.
And it is INCLUSIVE on both parts of the range. ILOC is EXCLUSIVE at the end like the range function.

```python
#Row 0 and ALL columns
cars.loc[0,:]
```

```python
#first three rows and then ALL COLUMNS
cars.loc[[0,1,2],:]
```

```python
#another way to do it. Please note that in this way it is inclusive on both sides. 
cars.loc[0:2,:]
```

```python
cars.loc[0:5,['Price','Initial_day','Final_Day','Quote_price']]
```

```python
#all rows and then only the price column. this way you receive a series. 
cars.loc[:,'Price']
```

```python
Frame = cars.loc[:,'Price'].to_frame()
Frame
```

```python
#all rows and only two columns. 
cars.loc[:,['Price','Initial_day']]
```

```python
#rows section limits are included
cars.loc[0:2,['Price','Initial_day']]
```

```python
#another to do it
cars.head(3).drop(['Final_Day','Quote_price'],axis=1)
```

Label means not the integer position, the name of the row and or column.

# iloc function

Last number in ILOC is EXCLUSIVE as range in both rows and columns, the first number of the range is INCLUSIVE in both rows and columns

```python
#last part of the range is not included. 
list(range(0,4))
```

```python
#last number in ILOC is exclusive as range, in both row and columns
cars_second.iloc[:,0:4]
```

```python
#last number in ILOC is EXCLUSIVE as range, the first number of the range is INCLUSIVE. 
cars_second.iloc[0:4,0:4]
```

```python
#loc is inclusive in the last part of the range. 
cars_second.loc[0:2]
```

```python
#iloc is exclusive in the last part of the range. 
cars_second.iloc[0:2,:]
```

```python
import numpy as np
```

```python
df = pd.DataFrame({
'name':['Jane','John','Ashley','Mike','Emily','Jack','Catlin'],
'ctg':['A','A','C','B','B','C','B'],
'val':np.random.random(7).round(2),
'val2':np.random.randint(1,10, size=7)
})
df
```

# 1. Logical operators
We can use the logical operators on column values to filter rows.

```python
df[df.val > 0.5]
```

```python
df[df.name > 'Jane']
```

```python
#or combination. 
df[(df.val > 0.5) | (df.name > 'Jane')]
```

```python
#and logic operation 
df[(df.val > 0.1) & (df.name > 'Jane')]
```

```python
#The logical operators also works on strings.
#Only the names that come after ‘Jane’ in alphabetical order are selected.
df[df.name > 'Jane']
```

# 2. Multiple logical operators

```python
df[(df.val2 > 1) & (df.val == 0.03)]
```

```python
df[(df.val < 0.5) | (df.val2 == 8)]
```

```python
#you can use it in strings, names more than C in alphabetical order. 
df[(df.val < 0.5) | (df.val2 == 8) | (df.name > 'C')]
```

```python
df[(df.val < 0.5) & (df.val2 > 2) & (df.name > 'C')]
```

The “&” signs stands for “and” , the “|” stands for “or”.

# 4. Str accessor

Pandas is a highly efficient library on textual data as well. The functions and methods under the str accessor provide flexible ways to filter rows based on strings.

```python
df[df.name.str.startswith('J')]
```

```python
df[(df.name.str.startswith('J')) & (df['val'] >0.02)]
```

```python
df[df.name.str.endswith('n')]
```

```python
df[df.name.str.contains('y')]
```

# 5. Tilde (~)

The tilde operator is used for “not” logic in filtering. If we add the tilde operator before the filter expression, the rows that do not fit the condition are returned.

```python
the tiklde always goes before the parenthesis. 
```

```python
df[~(df.name.str.startswith('J'))]
```

```python
df[(df.name.str.startswith('J')) & ~(df['val'] <0.02)]
```

```python
#not mless than 8
df[~(df['val2']<8)]
```

```python
df[~(df['name'].isin(['John','Emily']))]
```

# 6. Query

The query function offers a little more flexibility at writing the conditions for filtering. We can pass the conditions as a string.

```python
df.query('ctg == "B" and val > 0.5')
```

```python
df.query('ctg == "B" or val > 0.5')
```

```python
df.query('(ctg == "B") or (val > 0.5 and val2 >1)')
```

# 7. Nlargest or nsmalles

In some cases, we do not have a specific range for filtering but just need the largest or smallest values. The nlargest and nsmallest functions allow for selecting rows that have the largest or smallest values in a column, respectively.

```python
df.nlargest(3, 'val')
```

```python
df[['name','val']].nlargest(3, 'val')
```

```python
df.nsmallest(2, 'val2')
```

```python
df[['name','val']].sort_values('val')
```
