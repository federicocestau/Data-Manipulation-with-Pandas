# what if you want to join on one index. 

```python
movies_3 = movies.set_index('movieId')
movies_3
```

```python
#in the result it matches the left index with the right column but it uses the index from
#the left table in the result. 
movies_ratings_2 = pd.merge(movies_3,ratings, left_index=True,right_on='movieId').head()
movies_ratings_2
```

```python
movies_ratings_2.sort_index(ascending=True)
```

# What if you want to join on two indexes

```python
ratings_2 = ratings.set_index('movieId')
ratings_2
```

```python
#the index here does not have to be unique. 
movies_ratings_3 = pd.merge(movies_3,ratings_2,left_index=True,right_index=True)
movies_ratings_3
```

```python
#four types of joins: 
#1)inner join: use INTERSECTION OF KEYS FROM TWO DATAFRAMES. preserve the order of the left keys.
#there are no null values in join column. 
#2)outer join:Include all the keys from both tables. It is like full outer join in SQL.
#3)left join: use only keys from the left frame, similat to a SQL left outer join. Preserve key order. 
#4) right join: use only keys from the right frame, similar to SQL right outer join.Preserve key order.
```

There are four main types of joins: inner join, full outer join, left outer join and right outer join. 
The major difference between inner and outer joins is that inner joins result in the intersection of two tables, 
whereas outer joins result in the union of two tables.

An full outer join is a method of combining tables so that the result includes unmatched rows of both tables. If you are joining two tables and want the result set to include unmatched rows from both tables, use a FULL OUTER JOIN clause. The matching is based on the join condition.

```python
A = pd.DataFrame({'color':['green','yellow','red','white'],
                 'num':[1,2,3,4]})
A
```

```python
B = pd.DataFrame({'color':['green','yellow','orange','blue'],
                 'size':['S','M','B','SB']})
B
```

```python
B.second = pd.DataFrame({'color':['green','yellow','orange','blue'],
                 'measure':['H','J','K','L']})
B.second
```

```python
B.third = pd.concat([B,B.second], 
                  axis = 1)
B.third
```

```python
B.fourth = pd.DataFrame({'color':['green','yellow','orange','blue','red'],
                 'measure':['H','J','K','L','W']})
B.fourth
```

the concatenate function accepts null values

```python
B.seventh = pd.concat([B,B.fourth], 
                  axis = 1)
B.seventh
```

```python
B.fourth =  pd.concat([B,B.second], 
                  axis = 0)
B.fourth
```

Append function

```python
Head_table = pd.DataFrame({'Head':['blue','yellow','red','white'],
                 'num':[1,2,3,4]})
Head_table
```

```python
Head_table_2 = pd.DataFrame({'Type':['b','y','r','w','r','u'],
                 'values':[11,21,31,41,67,54]})
Head_table_2
```

```python
Head_3 =  pd.concat([Head_table,Head_table_2], 
                  axis = 0)
Head_3
```

```python
Head_table.append(Head_table_2)
```

```python
#Append is like concatenate with the axis=0.
B.append(B.second)
```
