# Combine the output of an agreggation in a dataframe

```python
#when you do a group by with an aggregation you receive less data compare to the orginal dataframe
```

```python
#to resolve this use the transform function
total_quantity = Invoice_numbers2.groupby('customerid')['quantity'].transform('sum')
```

```python
#total quantity by customer_id
Invoice_numbers2['total_quantity'] = Invoice_numbers2.groupby('customerid')['quantity'].transform('sum')
```

```python
Invoice_numbers2
```

```python
#with this method you get the percentage of the quantity of one row for the total quantity PER CUSTOMER_id in this case
#because the grouping was done that way
Invoice_numbers2['percentage_of_quantity_customer_id'] = Invoice_numbers2['quantity']/Invoice_numbers2['total_quantity']
```
